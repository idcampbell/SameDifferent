```python
import numpy as np
import pandas as pd
import torch
from generate_data import generate_sprites, generate_features

from utils import train, test, setup
```

### Convolutional model 

```python
def generate_mphate_examples(sprites, device, nx=4, nz=16, nepochs=1000):
    nitems = sprites.shape[0]
    task1 = torch.zeros([nitems, 2]).to(device)
    task2 = torch.zeros([nitems, 2]).to(device)
    task1[:, 0] = 1
    task2[:, 1] = 1
    xs = torch.zeros([nepochs, nitems*2, 4]).to(device)
    zs = torch.zeros([nepochs, nitems*2, 16]).to(device)
    return sprites.to(device), xs, zs, task1, task2 
```

```python
# Track the representations learned by the network during training.
nepochs = 1000
device = 'cuda'
colors_list = ['aqua', 'blue', 'fuchsia', 'green', 'grey', 'lime', 'maroon', 'navy', 'olive', 'purple', 'red', 'silver', 'teal', 'white', 'yellow', 'orange']
sprites, colors, shapes = generate_sprites('../sprites.npy', colors_list, len(colors_list))
mphate_sprites, xs, zs, task1, task2 = generate_mphate_examples(sprites, device, nepochs=nepochs)
```

```python
test_perfs = {}
for split in range(2, 14):
    # Model setup.
    model, train_loader, test_loader, train_dataset, test_dataset, optimizer, device = setup('data/image_comparisons.pt',
                                                                                             'data/image_metadata.csv',
                                                                                             train_cond='isTrain_split'+str(split)+'_all', 
                                                                                             encoder='conv')
    
    
    # Track the test and training loss of the networks during training.
    train_losses = torch.zeros([nepochs])
    test_losses = torch.zeros([nepochs])
    test_epochs = torch.zeros([int(nepochs/10)])
    
    # Train the network for nepochs.
    for i_epoch in range(nepochs):
        train_loss = train(model, device, train_loader, optimizer)
        if i_epoch % 10==0:
            test_loss, xs, zs = test(model, device, test_loader, mphate_sprites, task1, task2, xs, zs, i_epoch)
            test_losses[int(i_epoch/10)] = test_loss
            test_epochs[int(i_epoch/10)] = len(train_dataset)*i_epoch
        if i_epoch % 500==0:
            print(f'Epoch {i_epoch} train set average loss: {train_loss:.8f}')
            print(f'Epoch {i_epoch} test set average loss: {test_loss:.8f} \n')
        train_losses[i_epoch] = train_loss
    # Add the test and train performance.
    test_perfs['split'+str(split)] = (test_losses, test_epochs)
```

```python
import matplotlib.pyplot as plt

colors = ['maroon', 'darkorange', 'gold', 'green', 'lightblue', 'blue', 'purple', 'black']#['maroon', 'red', 'orangered', 'darkorange', 'gold', 'yellow', 'green', 'lightblue', 'blue', 'darkblue', 'mediumpurple', 'violet']
keys = ['split2', 'split3', 'split4', 'split5', 'split6', 'split7', 'split9', 'split10']
for i, split in enumerate(keys):
    plt.plot(test_perfs[split][1], test_perfs[split][0][:100], label=split, c=colors[i]) #
    if i==10:
        break
plt.xlim(-100000,6e6)
plt.ylim(0, 0.004)
plt.legend(loc='upper right')
plt.title('Image test accuracy')
plt.ylabel('Test set error')
plt.xlabel('Training epoch')
#plt.savefig('image-test-acc.png')
```

### Linear

```python
nepochs = 2000
# Track the representations learned by the network during training.
device = 'cuda'
sprites, colors, shapes = generate_features(16, 16) 
mphate_sprites, xs, zs, task1, task2 = generate_mphate_examples(sprites, device, nepochs=nepochs)
    
test_perfs = {}
for split in range(2, 8):
    # Model setup.
    model, train_loader, test_loader, train_dataset, test_dataset, optimizer, device = setup('data/linear_comparisons.pt', 
                                                                                             'data/linear_metadata.csv', 
                                                                                             train_cond='isTrain_split'+str(split)+'_all', 
                                                                                             encoder='linear')
    
    # Track the test and training loss of the networks during training.
    train_losses = torch.zeros([nepochs])
    test_losses = torch.zeros([int(nepochs/10)])
    test_epochs = torch.zeros([int(nepochs/10)])
    
    # Train the network for nepochs.
    for i_epoch in range(nepochs):
        train_loss = train(model, device, train_loader, optimizer)
        if i_epoch % 10 == 0:
            test_loss, xs, zs = test(model, device, test_loader, mphate_sprites, task1, task2, xs, zs, i_epoch)
            test_losses[int(i_epoch/10)] = test_loss
            test_epochs[int(i_epoch/10)] = len(train_dataset)*i_epoch
        if i_epoch % 1000 == 0:
            print(f'Epoch {i_epoch} train set average loss: {train_loss:.8f}')
            print(f'Epoch {i_epoch} test set average loss: {test_loss:.8f} \n')
        train_losses[i_epoch] = train_loss
    
    # Add the test and train performance.
    test_perfs['split'+str(split)] = (test_losses, test_epochs)
```

```python
import matplotlib.pyplot as plt

colors = ['red', 'orange', 'gold', 'green', 'blue', 'mediumpurple', 'violet']
for i, split in enumerate(test_perfs):
    plt.plot(test_perfs[split][0], label=split, c=colors[i])
plt.legend()
plt.title('One-hot model test accuracy')
plt.ylabel('Test set error')
plt.xlabel('Training epoch')
plt.savefig('linear-test-acc.png')
```
